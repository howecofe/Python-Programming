#### [Lab] 섭씨 온도를 화씨 온도로 변환

### 파일 모드(File Mode)


### 텍스트 파일 쓰기


#### write()

* `write()`를 이용하여 파일에 쓰기

```
f = open("file.txt", 'w')
f.write("Hello Python")
f.close()
```

```
!cat file.txt
```

#### writelines()

* `writelines()`를 이용하여 list를 파일에 쓰기

```
list = ["One\n", "Two\n", "Three\n"]
f = open("file.txt", 'w')
f.writelines(list)
f.close()
```

```
!cat file.txt
```

#### 표준 입력 → 파일 쓰기

* 표준 입력을 `input()` 함수를 통해 입력 받고, `write()` 함수를 통해 파일에 쓰기

```
text = input("입력: ")
f = open("file.txt", 'w')
f.write(text)
f.close()
```

```
!cat file.txt
```

```
text = [str(text + '\n') for text in input("여러 값 입력: ").split()]
f = open("file.txt", 'w')
f.writelines(text)
f.close()
```

```
!cat file.txt
```

```
x = ""
text = ""
while x != 'q':
  x = input("반복 입력: ")
  text += x + '\n'

f = open("file.txt", 'w')
f.writelines(text)
f.close()
```

```
!cat file.txt
```

### [Lab] 구구단 결과를 파일에 쓰기

* 구구단 결과를 파일에 쓰기

```
f = open("dan.txt", 'w')
for i in range(2, 10):
  for j in range(1, 10):
    f.write("{0} x {1} = {2}\n".format(i, j, i * j))
  else:
    f.write("\n")

f.close()
```

```
!cat dan.txt
```

#### readline()

* `readline()`를 이용하여 텍스트 파일을 라인 단위로 읽고 화면에 출력

```
f = open('anthem.txt', 'r')
while True:
  line = f.readline()
  if not line: break
  print(line)

f.close()
```

#### readlines()

* `readlines()`를 이용하여 텍스트 파일의 여러 라인을 읽고 화면에 출력

```
f = open("anthem.txt", 'r')
lines = f.readlines()
print(lines)
f.close()
```

#### read()

* `read()`를 이용하여 텍스트 파일을 읽고 화면에 출력

```
f = open("anthem.txt", 'r')
data = f.read()
print(data)
f.close()
```

#### tell()

* 파일 포인터 위치 변경

```
f = open("anthem.txt", 'r')
while True:
  print(f.tell())
  line = f.readline()
  if not line: break
  print(line)
```

#### seek()

* 파일 포인터 위치 이동

```
f = open("anthem.txt", 'r')
f.seek(46)
line = f.readline()
print(line)
f.seek(0)
line = f.readline()
print(line)
f.close()
```

### with 문

* 항상 파일을 `open()` 함수로 열고, `close()` 함수로 닫아야 하는 일을 자동으로 처리
* with문을 이용하면 with 블록 내에서 파일을 열고 벗어나면 파일을 닫음

```
f = open("file.txt", 'w')
f.write("Hello Python")
f.close()
```

```
with open("file.txt", 'w') as f:
  f.write("Hello Python")
```

### [Lab] 표준 입력으로 받은 내용을 파일로 쓰기

* 사용자로부터 표준 입력 받음
* 받은 내용을 계속 파일 쓰기

```
f = open("file.txt", 'w')
line = ""
while True:
  line = input()
  if not line: break
  f.writelines(line + '\n')

f.close()
```

### 이진 파일 복사

* 하나의 이진 파일을 읽어서 다른 이진 파일로 쓰기 (복사)

```
input = open("the-old-guard.jpg", "rb")
output = open("the-old-guard_copy.jpg", "wb")
while True:
  fp = input.read(1)
  if not fp: break
  output.write(fp)

input.close()
output.close()
```

## 디렉토리 및 파일 처리

### 디렉토리 생성

* 디렉토리와 파일을 다루는 다양한 함수를 제공하는 shutil와 os 라이브러리

```
import os
import shutil
os.mkdir('test')
os.mkdir('test/1')
os.mkdir('test/2')
os.mkdir('test/3')
```

### 디렉토리 및 파일 복사

* `shutil.copy()`: 파일 복사
* `shutil.copytree()`: 디렉토리 전체 복사

### 디렉토리 및 파일 확인

* `isdir()`: 디렉토리 존재 확인
* `isfile()`: 파일 존재 확인
* `exists()`: 디렉토리/파일 존재 확인

```
import os.path
print(os.path.isdir('test'))
print(os.path.isdir('test/1'))
print(os.path.isdir('test/4'))
print(os.path.isfile('the-old-guard.jpg'))
print(os.path.isfile('the-old-guard_copy.jpg'))
print(os.path.isfile('the-old-guard_copy2.jpg'))
print(os.path.exists('test2'))
print(os.path.exists('test2/1'))
print(os.path.exists('test2/4'))
```

### 디렉토리 목록 보기

* `os.walk()`: 디렉토리 목록 보기

```
import os
for dir_name, dir_list, file_names in os.walk('./'):
  for file_name in file_names:
    print(os.path.join(dir_name, file_name))
   # './': 현재 디렉토리
```

### 디렉토리 및 파일 삭제

* `os.remove()`: 파일 삭제
* `shutil.rmtree()`: 디렉토리 안에 모든 파일/디렉토리 삭제

```
import os
os.remove('the-old-guard_copy2.jpg')
os.remove('the-old-guard_copy.jpg')
```

```
!ls
```

```
import shutil
shutil.rmtree('test')
shutil.rmtree('test2')
```
